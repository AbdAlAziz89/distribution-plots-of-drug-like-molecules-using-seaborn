```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your CSV data into a DataFrame
df = pd.read_csv("10k_sample_29-38.csv")

# Set the style for better visualization
sns.set(style="whitegrid")

# Plotting normal distributions
plt.figure(figsize=(16, 12))

# Plot a normal distribution for each column
for i, col in enumerate(['MW', 'HAC', 'sLogP', 'HBA', 'HBD', 'RotBonds', 'FSP3', 'TPSA', 'QED'], start=1):
    plt.subplot(3, 3, i)
    sns.boxplot(df[col])
    #plt.title(col)
#plt.subplot(3, 3, 1)
#plt.boxplot(df['MW'])
#plt.title('MW')
# Adjust layout
plt.tight_layout()

# Show the plots
#plt.show()
#save plot
plt.savefig('box_plots.png')
```

```python
# Set the style for better visualization
sns.set(style="whitegrid")

# Plotting normal distributions
plt.figure(figsize=(16, 12))

# Plot a normal distribution for each column
for i, col in enumerate(['MW', 'HAC', 'sLogP', 'HBA', 'HBD', 'RotBonds', 'FSP3', 'TPSA', 'QED'], start=1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col], kde=True, color="skyblue", bins=30)
    plt.title(col)

# Adjust layout
plt.tight_layout()

# Show the plots
#plt.show()
#save plot
plt.savefig('hist_plots.png')
```
